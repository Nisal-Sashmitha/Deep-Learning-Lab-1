```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/ThulinaWickramasinghe/DL_LAB_01/main/Au_nanoparticle_dataset.csv'
df = pd.read_csv(url)
df.drop(['N_total','N_bulk','N_surface','R_avg'], axis=1)

df.head(20)
```

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
mean_values = new_df.mean()
std_values = new_df.std()
quartile_values = new_df.quantile([0.25, 0.5, 0.75])
```

```python
plt.figure(figsize=(15, 5))

for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_df[column], bins=20, edgecolor='black')
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_diag(sns.histplot, kde=True)  # Histograms with KDE
g.map_upper(sns.histplot, bins=20)  # Bivariate Histograms
g.map_lower(sns.kdeplot)  # Bivariate Kernel Density Plots

plt.show()
```
