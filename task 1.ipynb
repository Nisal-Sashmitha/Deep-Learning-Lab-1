```python
import matplotlib.pyplot as plt
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="orange", label="uniform")
plt.hist(np.random.randn(100000), density=True, bins=50, histtype="step", color="green", label="normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Plot of Z = X^2 + Y^2')
plt.show()

```

```python
import seaborn as sns
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Display first 5 observations
df.head()

corr_pearson = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')
corr_spearman = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.heatmap(corr_pearson, annot=True, cmap='coolwarm', center=0)
plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)
sns.heatmap(corr_spearman, annot=True, cmap='coolwarm', center=0)
plt.title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
